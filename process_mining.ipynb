## Introduction and Preliminaries

### Business Processes

Business processes are like a set of instructions that help a company run smoothly. These are steps that teams and individuals follow to get work done, much like a recipe. When you follow the steps, you can turn raw materials, like time and effort, into something valuable, like a product or a service that customers want. Good business processes are important because they help the company do things faster, make fewer mistakes, and save money. They also make it easier to see where improvements can be made. Just like how a good recipe can be tweaked and improved, businesses can also adjust their processes to become even better. 

### Process Modelling

Understanding and optimizing business operations is significantly augmented by the use of process models. These graphical representations make explicit the flow and interactions of tasks, illuminating the intricacies of each role. By visually mapping these procedures, it's possible to detect and address any potential bottlenecks, redundancies, or areas of wastage.

Process models come in different types, each offering a unique perspective. Workflow diagrams, for instance, illustrate task progression from one entity to another. Data flow diagrams trace the movement of data within the system, while business process modeling notation ([BPMN](https://www.bpmn.org/)) provides a standardized method for detailing business processes. You can see an example BPMN of simple payment process.

![Source) ](./images/process-model.webp)[Source](https://www.ibm.com/blog/bpmn/)

### Process Mining

Process Mining is a technology that addresses this need by turning vast amounts of event data into comprehensive, actionable knowledge.

At its core, Process Mining is a branch of business process management (BPM) and data science that uses event log data from an information system to analyze and visualize how processes operate in real life. It sits between data mining and BPM, bridging the gap between traditional statistical analysis and qualitative process analysis.

Imagine process mining as a traffic management system in a busy city. Each vehicle (process instance) follows a path (process) through the city's network of roads (activities). The traffic management system (process mining tool) observes the flow of vehicles, identifies traffic jams (bottlenecks), and spots cars taking long detours (deviations). With this information, it can adjust traffic light timings or suggest new routes (optimize the process), promoting a smoother and more efficient traffic flow. By analyzing the city's traffic patterns, process mining provides insights for making the city's transport system (business process) more efficient and effective.

Process Mining enables organizations to discover, monitor, and improve real processes by extracting knowledge from event logs readily available in today's systems. It can answer questions like:
- What are the most common paths in a process?
- Where are the bottlenecks in a process?
- What is the impact of one step on the subsequent ones?


### Process Mining Techniques and Tools

Process mining techniques are usually classified into three main categories: Discovery, Conformance, and Enhancement.
- **Discovery:** The most common technique, it takes an event log and produces a process model without any prior information. It allows you to visualize how processes are actually being performed.
- **Conformance:** This technique compares an existing process model with an event log of the same process. It checks if reality, as recorded in the log, conforms to the model and vice versa.
- **Enhancement:** It extends or improves an existing process model using information about the actual process recorded in some event log.

Various tools have been developed to implement these techniques. Prominent among them are:

- **Celonis:** An AI-driven platform that helps companies quickly visualize and evaluate their business processes. It offers a sophisticated interface and wide-ranging integration capabilities.
- **SAP Signavio** Process Intelligence: This tool enables businesses to understand and visualize their actual business processes. It brings process mining capabilities to SAP's suite, making it a suitable choice for businesses that already use SAP software. It's ideal for discovery, conformance checking, and enhancement of business processes.
- **ProcessGold:** A process mining tool that offers full-spectrum conformance checking and an easy-to-use interface.
- **Disco:** A user-friendly tool developed by Fluxicon, useful for quick and straightforward process mining analyses.

### Applications

It becomes extremely useful in scenarios where a multitude of objects of different types interact, such as:
- Healthcare: Managing patient flow, where the objects can be patients, doctors, equipment, medications, rooms, etc.
- Manufacturing: Tracking product lifecycle, where objects can be raw materials, machines, operators, maintenance tools, etc.
- E-commerce: Understanding customer behavior, where objects can be customers, products, webpages, customer service agents, etc.

- Retail: Streamlining inventory and sales forecasting, where the objects can be items for sale, store locations, suppliers, customers, sales associates, promotional events, etc.


### Event Logs
At the heart of process mining lies a crucial element - the event log. An event log can be described as a recorded sequence of activities that occurred within a system. It provides the raw data needed for process mining and serves as the basis for process discovery, conformance checking, and enhancement.

To understand event logs better, let's imagine a simple online shopping system. In this system, a customer's order might go through several stages - placing an order, payment processing, packing, shipping, and delivery. Each of these stages is an event that gets recorded in the event log.

For each event, the log typically contains several pieces of information:
- Case ID: This identifies the specific instance of the process. In our library system example, each individual book would have a unique case ID.
- Activity: This describes what happened in the event. It could be 'Book Checked Out', 'Book Renewed', 'Book Returned', etc.
- Timestamp: This records when the event took place. It allows us to reconstruct the sequence of activities for each case.
- Resource: This records who or what performed the activity. It could be a person (like the library member who checked out or returned the book) or a system component (like the library system that recorded the transaction and calculated the overdue fine).

An example of an event log from our library system might look something like this:

| Case ID | Activity                | Timestamp           | Resource |
|---------|-------------------------|---------------------|----------|
| 3001    | Book Checked Out        | 2023-04-01 10:30:00 | Member   |
| 3001    | Book Renewed            | 2023-04-14 09:15:23 | Member   |
| 3001    | Book Returned           | 2023-04-28 15:10:10 | Member   |
| 3001    | Overdue Fine Calculated | 2023-04-28 15:15:10 | System   |
| 3001    | Overdue Fine Paid       | 2023-04-30 11:20:44 | Member   |



By analyzing such event logs, process mining tools can reconstruct the actual process flow, identify bottlenecks, discover deviations from the ideal process, and suggest improvements. They provide a factual basis for understanding what happened in a process, when it happened, in what order, and by whom, removing guesswork and assumptions.

It's important to note that event logs can become quite complex in real-world scenarios. They might involve multiple interconnected processes, various resources, and a multitude of different activities. Therefore, understanding how to interpret and analyze event logs is a fundamental skill in process mining. In the following sections, we will delve deeper into the techniques and methods for event log analysis.



#### Event Log Formats
Event logs can be stored and represented in various formats. The most commonly used formats include XML, JSON, XES, and OCEL. Each of these formats has its own advantages and unique characteristics.

- **CSV (Comma Separated Values):** CSV is a simple file format that is widely supported by consumer, business, and scientific applications. Each line of the file is a data record and each record consists of one or more fields, separated by commas. CSV files can be easily read and written by many programs, although they are less flexible than other formats.

    Example of an event in CSV format:
```csv
    caseID,activity,timestamp,resource
    1001,Order Placed,2023-06-22 10:15:00,Customer
```

- **XML (eXtensible Markup Language):** XML is a markup language that defines a set of rules for encoding documents in a format that is both human-readable and machine-readable. It uses tags to define elements and can be used to represent complex data structures. However, XML is more verbose compared to other formats, which might lead to larger file sizes.

    Example of an event in XML format:
```xml
    <event>
        <caseID>1001</caseID>
        <activity>Order Placed</activity>
        <timestamp>2023-06-22 10:15:00</timestamp>
        <resource>Customer</resource>
    </event>
```

- **JSON (JavaScript Object Notation):** JSON is a lightweight data-interchange format that is easy for humans to read and write and easy for machines to parse and generate. It uses a key-value pair structure that is similar to how data is represented in many programming languages, making it a popular choice for data exchange.
	
    Example of an event in JSON format:
```json
    {
        "event": {
            "caseID": 1001,
            "activity": "Order Placed",
            "timestamp": "2023-06-22 10:15:00",
            "resource": "Customer"
        }
    }
``` 

- **XES (eXtensible Event Stream):** XES is a standard specifically developed for process mining applications. It provides a way to store, manage, and exchange event data while retaining semantic integrity. XES allows storing additional attributes for events, cases, and logs, enabling rich event log data.
	
    Example of an event in XES format:
```xml
    <trace>
        <string key="concept:name" value="1001"/>
        <event>
            <string key="concept:name" value="Order Placed"/>
            <date key="time:timestamp" value="2023-06-22T10:15:00.000+00:00"/>
            <string key="org:resource" value="Customer"/>
        </event>
    </trace>
```

- **OCEL (Object-Centric Event Logs):** OCEL is a recent extension to the XES standard, designed specifically for Object-Centric Process Mining. It allows multiple types of events related to different objects to be recorded in a single log, enabling a more comprehensive understanding of processes involving multiple interacting objects.
	
    Example of an event in OCEL format:

```xml
    <log>
        <global>
            <string key="concept:name" value="Online Shopping"/>
        </global>
        <objects>
            <object id="order_1001">
                <string key="concept:name" value="Order"/>
                <string key="concept:type" value="Case"/>
            </object>
            <object id="customer_1">
                <string key="concept:name" value="Customer"/>
                <string key="concept:type" value="Resource"/>
            </object>
        </objects>
        <events>
            <event id="event_1">
                <string key="concept:name" value="Order Placed"/>
                <date key="time:timestamp" value="2023-06-22T10:15:00.000+00:00"/>
                <object-ref id="order_1001"/>
                <object-ref id="customer_1"/>
            </event>
        </events>
    </log>
```

In this example, there is one event - 'Order Placed'. This event involves two objects - 'order_1001' and 'customer_1'. By using object-ref tags, we can indicate that both these objects were involved in the event.

The 'order_1001' object has the concept:type of 'Case', indicating that it's the central object of interest, while 'customer_1' has the concept:type of 'Resource', indicating that it's the entity that performed the activity.


In summary, each event log format brings unique strengths and potential limitations to the table. CSV is straightforward and universally supported, making it excellent for simple, tabular data but falls short in representing complex hierarchical data structures. XML is comprehensive and great for complex structures, yet can be verbose and lead to larger file sizes. JSON is lightweight and easy to use, but like CSV, is less suitable for very complex data structures. XES is tailor-made for process mining and offers semantic integrity, allowing it to handle complex event log data and additional attributes, although it might require a steeper learning curve for users unfamiliar with process mining. Lastly, OCEL, an extension of XES, allows for the logging of interactions between multiple objects, which is advantageous for understanding complex processes but may introduce additional complexity.

### Converting Event Logs

converts a given DataFrame, event stream, or log to the event log format, which is structured format for process mining analyses.

And there are many more transformation options that can be checked [here](https://pm4py.fit.fraunhofer.de/static/assets/api/2.7.5/generated/pm4py.convert.html)

#### Event Log Flattening

Flattening event logs is a very important concept in process mining. Object-centric event logs, in particular OCEL files, can be flattened to XES to apply traditional process mining approaches. In the transformed log, the entities belonging to a specific entity type serve as the cases, with each case comprising all events associated with that particular entity.
PM4PY offers `ocel_flattening` for this operation. As an example:

````python
ocel_log = pm4py.read_ocel("data/input/order_log.jsonocel")
traditional_logs = pm4py.ocel.ocel_flattening(ocel_log, "order")
traditional_logs.head(5)
````

PM4PY also allows you to precheck the validity of the format the OCEL file. This can be achieved by either [jsonocel.apply](https://pm4py-source.readthedocs.io/en/stable/_modules/pm4py/objects/ocel/validation/jsonocel.html) or [xmlocel.apply](https://pm4py-source.readthedocs.io/en/stable/_modules/pm4py/objects/ocel/validation/xmlocel.html) as the following:
```
    jsonocel.apply(<OCEL_FILENAME>, <OCEL_SCHEMA>)
```

### Process Trees

A process tree provides a structured, hierarchical depiction of processes where nodes represent either activities or control-flow constructs.

#### What are Process Trees?
- **Activities**: These are the tasks or events that are executed as part of the process.
- **Operators**: They define control-flow semantics. Common operators include:
    - **→ (Sequence)**: Indicates that activities occur one after the other.
    - **× (Exclusive Choice)**: Denotes a decision point where only one path is taken.
    - **∥ (Parallel)**: Implies that multiple activities can occur simultaneously.

#### Building and Visualizing a Process Tree

````python
log = pm4py.read_xes('data/input/running-example.xes')
process_tree = pm4py.discovery.discover_process_tree_inductive(log)

from pm4py.visualization.process_tree import visualizer as pt_visualizer
tree_view = pt_visualizer.apply(process_tree)
pt_visualizer.view(tree_view)
````

### Petri Nets

A powerful tool for modeling concurrent systems, Petri nets consist of places, transitions, and tokens.

#### **What are Petri Nets?**
- **Places**: Represent conditions or states.
- **Transitions**: Denote events or activities. They can consume or produce tokens from/to places.
- **Tokens**: Sit in places and represent the potential for activities (transitions) to occur. For a transition to fire, specific conditions (tokens in certain places) must be met.

#### **Discovering and Visualizing a Petri Net in PM4PY**

````python
petri_net, im, fm = pm4py.convert.convert_to_petri_net(process_tree)
pm4py.vis.view_petri_net(petri_net)
````

### Directly-Follows Graphs (DFGs)

DFGs capture the essence of which activities directly succeed others in the event log.

#### What are Directly-Follows Graphs?
- Nodes in a DFG represent activities.
- Directed edges between nodes depict the succession relationship. The weight (or thickness) of an edge indicates the frequency of the observed succession.

#### Creating and Visualizing a DFG in PM4PY

````python
dfg, start_activities, end_activities = pm4py.discovery.discover_dfg(log)
pm4py.vis.view_dfg(dfg, start_activities, end_activities)
````
